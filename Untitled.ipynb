# Linux System Overview

# Operating System In a Nutshell

Programs are executed by a CPU that follows the Von Neumann fetch–decode–execute cycle.

An operating system manages hardware resources to ensure correct, efficient, and user-friendly execution of multiple programs. The techniques that is used to ensure easy of use are:
- **Virtualization**
- **Concurrency**
- **Persistence**

## Virtualization - General Scope
- OS takes a physical resource and transforms it into a more general, powerful, and easy-to-use virtual form of itself
- The OS provides each program with the illusion of running on its own virtual machine
- **System calls** are used to access features of the virtual machine, such as running program, allocating memory or accessing file.
- Programs usually call library wrappers (e.g., libc), and these wrappers internally perform the system call
- Virtualization allow many program to run and have shared access to hardware, it needs to manage these resouces to ensure program runs safely and efficiently, therefore, sometimes operating system is known as resource managers
- An OS **virtualizes the CPU**, so each process appears to run independently. To achieve this, in the background, OS is time-sharing the CPU
- Each process has it's own **virtual address space**, so when two processes are running simultanously, it might appear that they're modifying the same memroy space when in fact they're modifying different memory address in actual memory

### Example
| Physical | Virtual Abstraction |
|:---|:---|
| CPU | Process/thread |
| Memory | Virtual address space |
| Disk | File system |
| Device | File descriptor |

## Concurrency
- Multiple tasks make progress during the same time period, it improves responsiveness, throughput, and resource utilization
- Concurrency is achieved via time-slicing on one CPU or parallel execution on multiple CPUs/cores
- The OS rapidly context-switches between workers known as **threads** to create the illusion of simultaneous execution (on a single CPU)
- **thread** is the smallest unit scheduled by the CPU. It has it's own **program counter**, **register**, **stack**, and shared **code**, **heap**, and **files** with other threads in the same process
- **Multi-threaded** programs allow better performance, responsiveness and lower overhead than multiple processes
- Race conditions occur because operations can be interleaved unless protected by **atomic execution** (indivisible/cannot be interrupted) or synchronization.

### Process vs Thread
- Process -> independent program with its own memory space, the example includes:
    - Browser
    - Text editor
    - Terminal
- Process has it's own:
    - Address space (memory)
    - Heap
    - Stack(s)
    - Open files
    - OS metadata (PCB)
- Thread -> lightweight execute unit inside a process, sharing memory, example includes:
    - UI thread
    - Network thread
    - Rendering thread
- Thread shares:
    - Code
    - Heap
    - Global variables
    - Files
- Thread has it's own:
    - Program counter
    - Registers
    - Stack

## Persistence
- Data remains stored safely and correctly on non-volatile storage (disk/SSD) even after crashes, reboots, or power loss
- Without it, memory is lost on shutdown and files would disappear
- The OS still provides private virtual memory for processes, while persistent storage is shared through the file system with access control
- This can be achieved by storing data in non-volatile memory such as SSD
- OS assumes user want to share non-volatile memory, therefore it does not create private and virtualized memory
- System calls such as `open()`, `write()` and `close()` are used to store data in non-volatile memory
- Hardware devices are accessed through **device drivers** inside the OS kernel, while user programs access services via system calls (often through **standard library** wrappers).
- Techniques further used to provide persistence
    - **Journaling** -> records changes in a log before applying them to main file
    - **Copy-on-Write** -> delays copying data until it's modified

# Design Goals

In the simplest term, the purpose of operating system is to provide:
- Abstraction
- Improved performance by minimizing overhead
- Protection between applications by isolating processes from one another
- Reliability
- Energy efficiency
- Security
- Mobility by allowing it run on many devices

## Kernel & Kernel Architecture
A kernelized system is an OS design where a privileged kernel runs in protected mode and directly manages hardware, while user programs run in user mode and access services via system calls.
To achieve this, the system is split into:
- Kernel space (privileged) → hardware control, protection, core services
- User space (unprivileged) → applications
The goal is to provide:
- safety (fault isolation)
- security (restricted access)
- stability (user bugs can’t crash hardware directly)
Kernels responsibilities include:
- CPU scheduling
- memory management
- file systems
- device drivers
- system calls / IPC
- protection & security

## Monolithic Kernel
All OS services run inside one large kernel address space.
Everything executes in kernel mode:
- scheduler
- memory manager
- file systems
- drivers
- networking

Pros:
- very fast (direct function calls)
- low overhead
- simple communication

Cons:
- poor isolation
- one bug can crash whole system
- large trusted code base
- harder to maintain

Examples:
- Linux
- Windows (hybrid but mostly monolithic-like)

## Microkernel
Only minimal core services stay in kernel, and everything else runs in user space servers.

Kernel keeps only:
- scheduling
- memory management
- IPC

Other services become processes:
- file system server
- driver servers
- network stack

Pros:
- strong isolation (crashes contained)
- better security
- easier debugging
- modular

Cons:
- slower (IPC/message passing overhead)
- more complex design

Examples:
- Minix 3
- QNX
- macOS (hybrid using Mach microkernel ideas)

## OS Structuring Models & Interaction Styles
Kernel interaction still happens through:
- system calls
- traps: a controlled transfer of execution from user mode to kernel mode
- IPC: allows separate processes to communicate and coordinate

**Client-Server Model**
- OS services are structured as servers, and programs act as clients that request services
- The path can be described as: client → sends request → server → performs service → reply returned
- Processes communicate via IPC, message passing and system calls
- Benefits:
    - modular
    - fault isolation
    - easier debugging
    - replaceable components
- Cost: IPC overhead → slower than direct calls
 
**Object-Oriented Model**
- OS components are structured as objects with data, methods and well-defined interfaces
- Benefits:
    - clean interfaces
    - maintainability
    - modularity
    - safer abstraction boundaries

**Process Hierarchy Mode**
- Processes are organized in a parent–child tree structure. The process is typically: parent creates child (fork/spawn) → child inherits resources → parent may manage or wait for child.
- It used heavily in Unix-like systems in the form of **shells spawn commands** and **servers spawn workers**
- Benefits
    - structured control
    - resource inheritance
    - easier cleanup
    - natural job control

## Mechanism vs Policy in Operating Systems
**Mechanism** -> implementation tools or procedures
Properties:
- general
- reusable
- policy-independent
- rarely changed
Examples:
- context switch code
- page tables
- system calls
- interrupt handling
- IPC primitives
- timers
- Scheduler mechanism:
    - Ability to save/restore registers
    - Maintain ready queue
    - Switch processes

**Policy** -> decision rules
Properties:
- higher-level logic
- changeable
- workload-dependent
- often tunable
Examples:
- Which process runs next (FCFS, RR, priority)
- Which page to evict (LRU, FIFO)
- How much memory to allocate
- Access control rules

**Symbiosis Example**

- Mechanism → paging hardware + page tables

File System
- Mechanism → read/write blocks
- Policy → caching strategy

# System Stack

A stack used by the OS kernel to handle privileged execution such as system calls, traps, interrupts, and context switches.

When a program runs:
- User mode → uses user stack
- Kernel mode → uses system (kernel) stack
This separation protects the kernel and ensures safe execution.

A separate system stack is needed to ensure:
- Security
    - User stack is untrusted
    - Kernel must not rely on user memory
    - Prevents corruption or attacks
- Stability
    - Kernel operations must have guaranteed safe stack space
- Isolation
    - User bugs cannot damage kernel state

**System Call Flow**:
```
User program calls read()
    ↓
Trap instruction
    ↓
CPU switches to kernel mode
Switches to system stack
    ↓
Kernel runs on system stack
    ↓
Return → back to user stack
```

Typical contents in system/user stack:
- saved registers
- return addresses
- local kernel variables
- trap/interrupt frames
- system call parameters
- context switch data

# Process

- A process is a program in execution, not just a file on disk.
- It represents the active state of a running program, including its memory, CPU state, and resources.
- It is the basic unit of scheduling and resource ownership in an operating system.

## Why processes exist (multiprogramming & concurrency)
- A key OS goal is multiprogramming: allowing users to run multiple programs simultaneously.
- The OS achieves this by creating concurrency through:
    - running one process for a short time slice
    - saving its state
    - switching to another process
- This technique is called time-sharing or context switching.
- Even on a single CPU, this creates the illusion of parallel execution.
- On multi-core systems, processes may actually run in parallel.

Often a process consist of:
**Memory state (address space)**
- Code (text segment) – program instructions
- Data segment – global/static variables
- Heap – dynamic memory allocation (malloc/new)
- Stack – function calls, local variables
- Virtual address space – private memory view of the process

**CPU execution state**
- Program Counter (PC) – next instruction to execute
- Registers – general-purpose CPU values
- Stack Pointer (SP) – top of stack
- Frame Pointer (FP) – function frame reference
- Instruction Pointer (IP) – often same as PC (architecture-dependent)
These must be saved/restored during context switches.

## OS-managed resources
- Open files (file descriptors)
- I/O devices
- Network sockets
- Signal handlers
- Security/permissions info
- Scheduling info (priority, state)

## Process states
Most OSs track processes using states:
- New
- Ready
- Running
- Waiting/Blocked (I/O)
- Terminated
This helps the scheduler decide which process runs next.

## Process API (system calls)
The OS provides system calls to manage processes:

**Lifecycle operations**
- Create – start a new process (fork, spawn, exec)
- Destroy – terminate a process (exit, kill)
- Wait – wait for a child to finish (wait, waitpid)

**Control operations**
- Suspend/resume
- Change priority
- Send signals
- Allocate memory
- Manage resources

**Status operations**
- Get process ID (PID)
- Check exit status
- Query resource usage

## Process Creation
1. The OS loads or maps the program’s code and data into the process’s virtual address space, either eagerly or lazily using demand paging.
2. Sets up the process address space (code, data, heap, stack segments).
3. Allocate one or more stacks (one per thread) for function calls, local variables, and return addresses.
4. Reserve heap space for dynamic allocation, which user libraries manage using functions like `malloc()`.
5. Initializes CPU state (registers, program counter, stack pointer).
6. Creates a Process Control Block (PCB) and assigns a PID.
7. Initalize **file descriptor** for input(0)/output(1)/error(2)
8. Marks the process ready for scheduling.

# Process API & System Calls

System calls are the programming interface to operating systems such as Unix.
```
User Program
   ↓
System Calls  ← interface
   ↓
Kernel (Unix/Linux)
   ↓
Hardware
```

System calls form the interface between user programs and the kernel. They enable safe kernel access, manage resources, simplify programming through abstractions, enforce security, and provide communication and networking capabilities:

**User–Kernel Boundary**
- System calls provide a controlled entry point from user mode to kernel mode.
- User programs cannot execute privileged instructions directly.
- A system call triggers a trap, switching the CPU to kernel mode.
- The kernel safely performs the requested operation, then returns to user mode.
- This enforces protection and isolation between applications and hardware.
- Example: `read()`, `write()`, `fork()`

**Resource Management**
- The OS owns and manages hardware resources:
    - CPU
    - memory
    - disks
    - devices
- System calls allow programs to request and release resources.
- The kernel ensures:
    - fair sharing
    - scheduling
    - allocation
    - cleanup
- Examples:
    - memory -> `mmap()`, `brk()`
    - files -> `open()`, `close()`
    - processes -> `forck()`, `exit()`
**Streamlined Development**
- Provides high-level abstractions so programmers don’t deal with hardware details.
- Simplifies development:
    - read files without disk logic
    - create processes without CPU control
- Improves portability across POSIX systems.
- Example: `read(fd, buf, 100)`
**Security and Access Control**
- The kernel validates every system call.
- Checks:
    - permissions
    - ownership
    - capabilities
    - isolation rules
- Prevents:
    - illegal memory access
    - unauthorized file access
    - device misuse

- Examples:
    - file permission checks on `open()`
    - process isolation
    - `setuid()` rules

# C Programming Interface

Systems such as Unix is written in C, and designed to be programmed in C.

The C programming interface provides user-level wrappers around system calls, allowing applications to safely and portably access kernel services. It serves as the bridge between programs and the operating system.

## Characterstics of C System Calls
**Return value convention**:
- success -> non-negative value
- failure -> -1
- Example:
    - `int fd = open("file.txt", O_RDONLY); if (fd == -1) { /* error */ }`
    - `ssize_t n = read(fd, buf, 100); if (n == -1) { /* error */ }`
**Errors reported through errno**:
- When a call fails -> `return -1`, `set errno`
- Example:
    - `if (open("x", O_RDONLY) == -1) { perror("open") }`
- `errno` might be: `ENOENT`, `EACCES`, `EBADF`
- `errno` is thread-local so other process concurrently don't overwrite it
**Thin wrapper design**
- C system calls are usually:
    - very small wrappers
    - minimal overhead
    - just set registers + trap
- Example Flow: `read() → libc wrapper → syscall instruction → kernel`
**Blocking by default (synchronous)**
- Most system calls block until done, examples: `read()`, `wait()`, `accept()`
- You can request non-blocking with flags: `O_NONBLOCK`
**Simple types only**
- System calls use: int, pointers, structs
- NOT: complex objects, exceptions, dynamic allocation
- Reason: portability, language neutrality, minimal ABI complexity
**Explicit error checking required**
- C doesn’t use exceptions.
- So programmer must check: `if (read(fd, buf, n) < 0)`
## C System call execution sequence
```
User space
-----------
set registers
   ↓
syscall instruction
   ↓ (hardware switches privilege)
Kernel space
-----------
handle syscall
set return value
   ↓
return to user
```

## System call Tracking & Debugging
`strace` is one of the most widely used and often the first tool for debugging system-call and OS interaction issues. It is simple and powerful for file, process, and I/O problems, but other tools like `gdb` or `perf` are better for logic or performance debugging.
**observability**
```
Application logic      → gdb
Library/API behavior   → ltrace
System calls (kernel)  → strace
Kernel internals       → perf / ftrace / eBPF
Hardware performance   → perf
```

# Process & Scheduling

## Steps needed for a process to run
A process is a program in execution: code + current CPU state + memory + OS-managed resources.

1. Program exists on storage
- You have an executable file (e.g., ELF on Linux) on disk/SSD.
- It contains:
    - machine code and data sections,
    - metadata (entry point, required libraries, relocation info),
    - permissions.

2. A parent requests creation (e.g., fork() + exec())
- System like Unix create process via: 
    - `fork()`
        - OS creates a new process (child) that is (initially) a near-copy of the parent.
        - Modern Unix uses copy-on-write: pages are shared until one side writes.
    - `execve()`
        - Child replaces its memory image with the new program.
        - Loads executable segments into memory, sets up stack/argv/envp.
        - Resolves dynamic libraries (loader/linker).
- Alternative: some OS designs use “spawn” directly; Unix historically uses fork+exec.

3. OS creates kernel data structures (At minimum)
- PID assigned.
- Process Control Block (PCB) created (details later).
- Kernel allocates and initializes:
    - address space structures (page tables),
    - kernel stack,
    - file descriptor table references,
    - scheduling info.

4. Memory is prepared
- Virtual address space created:
- text/code segment (often read-only),
    - data + bss,
    - heap,
    - user stack,
    - memory-mapped regions (shared libs, mapped files).
- Page tables set and protection bits applied (R/W/X, user/supervisor).

5. Process becomes “ready”
- The process is placed into a ready queue (eligible to run).
- Not running yet until the scheduler selects it.

6. Scheduler dispatch + CPU state load
- Scheduler picks a ready process.
- CPU context is loaded (registers, PC instruction pointer, stack pointer).
- CPU is placed into user mode and begins at the program entry point.

7. It runs, blocks, yields, or exits
- run on CPU,
- block (e.g., waiting for disk/network/I/O),
- be preempted by timer interrupt,
- exit (return code, resources reclaimed).

## How multiple processes are run (FCFS, Round Robin, process switch)
A single CPU core can execute one process at a time, but the OS creates the illusion of concurrency by time-sharing.

1. FCFS (First-Come, First-Served)
Idea: Run processes in order of arrival until they finish or block.
- **Non-preemptive**: once a process has the CPU, it keeps it until it blocks/exits.
- Pros:
    - simple,
    - low overhead.
- Cons:
    - “convoy effect”: a long CPU-bound job makes short interactive jobs wait a lot.
- Where you see it:
    - simple batch systems,
    - sometimes as a baseline policy in teaching OS.
Flow:
- Processes enter ready queue.
- Scheduler picks head of queue.
- Runs until completion or blocking.
- Next process.

2. Round Robin (RR)
Idea: Each ready process gets a fixed time quantum (time slice). If it doesn’t finish, it goes to the back of the queue.
- **Preemptive**: timer interrupts enforce the quantum.
- Great for interactive systems if:
    - quantum is “small enough” to keep UI responsive,
    - but not too small (context-switch overhead grows).
- Typical flow:
    - pick next ready process,
    - run for at most Q milliseconds,
    - if it used all Q and still runnable → preempt → enqueue at tail,
    - if it blocked early → removed from CPU and waits on event.

### Process Switch
When the OS stops running one process and starts another, it performs a context switch (detailed later). This happens when:
- the running process blocks (I/O wait),
- the time slice expires (RR preemption),
- a higher-priority process becomes ready (priority scheduling),
- the process yields voluntarily,
- an interrupt handler wakes a process and scheduling decides to switch.

## How does a process offer protection from other processes?
This is process isolation: one process should not be able to read/write another process’s memory or hijack its CPU privileges.

1. Dual-mode operation: user mode vs kernel mode
- User mode:
    - cannot execute privileged instructions,
    - cannot directly access hardware,
    - restricted memory access.
- Kernel mode:
    - OS can do everything (privileged).

System calls are controlled gateways from user → kernel.

2. Virtual memory + hardware-enforced address spaces
- Each process gets a virtual address space.
- The CPU’s MMU translates virtual → physical addresses using that process’s page tables.
- Page tables include protection bits:
    - user vs supervisor,
    - read/write/execute permissions.
- If a process tries to access memory it doesn’t own → page fault → OS kills it or signals it (e.g., SIGSEGV).

3. Controlled sharing
- Processes can share:
    - shared memory segments,
    - memory-mapped files,
    - shared libraries,
- (…but only if the OS configures mappings intentionally with permissions.)

4. OS resource permissions
- Processes also are isolated by:
    - file permissions (UID/GID, ACLs),
    - capabilities (Linux),
    - namespaces / containers,
    - IPC permissions.

So “process protection” is primarily hardware memory protection + privilege separation, enforced by the OS.

## Common Process Queue
A queue is a kernel-managed list of processes waiting for something, which contains:
1. Ready queue
- Processes that are runnable (have everything except CPU).

2. Waiting / blocked queues
- Processes waiting for an event:
    - disk I/O completion,
    - network packet,
    - lock release,
    - child process exit,
    - timer expiry. (Often there are multiple wait queues, one per device or per synchronization object.)

3. New queue (conceptual)
- processes being created / admitted.

4. Terminated queue (conceptual)
- zombies awaiting parent wait() in Unix.

Schedulers typically manage:
- one or more ready queues (possibly per priority level),
- many wait queues.

## Time-runout detection by clock interrupt
In preemptive systems (like round robin), the OS needs to ensure no process runs forever.

**Mechanism: periodic timer interrupt**
- Hardware timer (programmable interval timer / local APIC timer) is set to interrupt every tick or after a programmed interval.
- When the interrupt fires:
    - CPU stops current user code,
    - enters kernel interrupt handler,
    - OS increments time/accounting,
    - decrements remaining quantum for the running process.

**Time slice expiration**
- If the running process used up its quantum:
    - mark it “ready” again,
    - move it to tail of ready queue,
    - schedule a different process.
This is the core enforcement for time-sharing.

## Context Switch
A context switch is saving the CPU state of one process/thread and restoring the CPU state of another.

**What must be saved/restored**
- General purpose registers (R0..Rn)
- Program counter / instruction pointer
- Stack pointer
- CPU flags/status register
- Possibly floating-point/SIMD registers (often lazily)
- Memory management context:
    - page table base register (e.g., CR3 on x86)
    - TLB effects (may flush or use tagged TLB/ASIDs)
- Kernel bookkeeping:
    - current process pointer,
    - scheduling stats,
    - kernel stack pointer.

**Why it’s expensive**
- Pure overhead: no user work gets done while switching.
- Cache/TLB effects can slow the new process after the switch.

## What does hardware do when an interrupt occurs?
At a high level (typical modern CPU behavior):
1. Finish/abort the current instruction boundary (architecture dependent).
2. Save minimal CPU state automatically:
    - at least the current instruction pointer (return address),
    - flags/status,
    - sometimes privilege level info.
3. Switch to a privileged mode (kernel/supervisor).
4. Switch to an interrupt stack (or kernel stack) if transitioning from user mode.
5. Lookup the interrupt vector in an interrupt descriptor table.
6. Jump to the interrupt handler (kernel code).

Then the OS handler runs, does work, and eventually returns via an interrupt return instruction, restoring the saved state.

## Process Control Block (PCB)

The PCB is the OS’s per-process record: everything the kernel needs to manage a process.

Typical fields:
- Identification
    - PID, parent PID, process group, session ID
- Process state
    - running/ready/blocked/suspended/terminated
- CPU context
    - saved registers, PC, SP, flags
- Scheduling info
    - priority, policy, time slice remaining, CPU usage
- Memory management info
    - pointer to page tables / address space descriptor
    - memory regions (text/data/heap/stack/mmap)
- I/O and files
    - open file table pointer, current working directory, root directory
- Credentials and security
    - UID/GID, capabilities, security labels (SELinux/AppArmor)
- Signals and IPC
    - signal masks, pending signals
    - IPC endpoints, shared memory descriptors
- Accounting
    - CPU time used, start time, resource limits.

Linux calls the main structure task_struct (plus mm_struct, files_struct, etc. attached).

## Unix system control blocks

Textbook “system control blocks” are the OS-wide structures that manage the system as a whole (as opposed to per-process PCBs). Typical examples:

1. Memory management control structures
    - free frame lists / buddy allocator structures
    - page frame descriptors
    - global VM accounting, swap management
    - kernel page tables

2. I/O system control blocks
    - buffer cache / page cache metadata
    - device driver tables
    - interrupt vector tables
    - per-device request queues

3. File system control blocks
    - superblocks (per mounted filesystem)
    - inode tables / vnode layer
    - global file table (open file descriptions)
    - mount table

4. Process management control blocks (global)
    - process table / task list
    - run queues / scheduling classes
    - PID allocator structures

Exact names differ by Unix flavor, but the concept is: global kernel state + tables supporting process, memory, and I/O management.

## Process Operation

1. CPU instruction cycle (micro view)
While running, the CPU repeats:
- fetch instruction at PC
- decode
- execute
- update PC and registers
The process “runs” until something causes a transition.

2. Process state transitions (macro view)
Common states:
- New: being created
- Ready: waiting for CPU
- Running: executing on CPU
- Blocked/Waiting: waiting for I/O/event
- Terminated: finished (may linger as zombie in Unix)
- Suspended (optional model): swapped out or stopped

Typical transitions:
- Ready → Running: dispatch
- Running → Ready: preempt (time slice) or yield
- Running → Blocked: waiting for I/O / lock / event
- Blocked → Ready: event completes (interrupt wakes it)
- Running → Terminated: exit
- Running → Suspended: stop signal / debugging / swapping
- Suspended → Ready: resume signal / swapped back in

3. Unix-specific “stop/resume”
- SIGSTOP, SIGTSTP: stop a process (state becomes stopped)
- SIGCONT: resume

Swapping out (old-style) is less common today, but “suspended” still exists conceptually via stopped state or cgroup freezer.

## What happens before the process: power on → hardware bootstrap → OS bootstrap

1. Power on
- CPU resets into a known state.
- Starts executing from a fixed address (reset vector).

2. Firmware stage (hardware bootstrap)
Depends on platform:
- BIOS (legacy) or UEFI (modern PCs)
- Firmware:
    - initializes basic hardware (memory, buses, devices),
    - performs POST,
    - selects boot device,
    - loads the bootloader.

3. Bootloader stage
Examples: GRUB, systemd-boot, LILO (older).
- Loads OS kernel image into memory.
- Loads initramfs/initrd if used.
- Passes kernel parameters (boot args).
- Jumps to kernel entry point.

4. Kernel bootstrap (OS bootstrap)
Kernel performs:
- CPU mode setup (paging, protected mode / long mode).
- Initializes:
    - interrupt descriptor tables and handlers,
    - scheduler and run queues,
    - memory allocator + VM subsystem,
    - device drivers (early),
    - root filesystem (often via initramfs first).

- Creates the first user-space process:
    - historically /sbin/init
    - on Linux often systemd as PID 1.

At this point, “process world” begins.

## Unix system startup: from init PID 1 to GUI

This differs a lot by Unix flavor and by init system. The most common modern Linux pattern (systemd) and note the classic SysV idea are:

1. PID 1 responsibilities
PID 1 is special because:
- it is the ancestor of (almost) all processes,
- it adopts orphaned processes,
- it is central to system startup and service supervision.

2. systemd startup (common on modern Linux)
- Kernel starts PID 1 = /sbin/init → systemd
- systemd:
    - mounts essential pseudo-filesystems: /proc, /sys, /dev (via udev)
    - starts device manager (udevd or integrated)
    - reads unit files and dependencies
- Brings system to a target (like runlevel):
    - multi-user.target (text mode, network services)
    - graphical.target (GUI)

During this:
- networking services start (NetworkManager/systemd-networkd),
- logging (journald),
- time sync, dbus, etc.

3. Login + display stack to start GUI
For GUI, typical chain:
- systemd reaches graphical.target
- starts a display manager (DM), e.g.:
    - GDM (GNOME), SDDM (KDE), LightDM, etc.
- DM starts the display server/compositor:
    - Wayland compositor (modern default on many distros) or
    - Xorg/X11 (older/compat)
- DM presents a login screen.
- After login, it starts the user session:
    - desktop environment (GNOME Shell, KDE Plasma, etc.)
    - per-user services (systemd --user), settings daemons, panel, etc.

4. Classic SysV init (historical model)
- PID 1 = init
- reads /etc/inittab
- enters a runlevel (0-6)
- executes scripts in /etc/rc.d/ or /etc/init.d/
- GUI typically started by:
    - runlevel 5 (on some systems) or
    - starting a display manager from an init script
```
init
 ├─ getty
 │   └─ login
 │       └─ bash
 │           ├─ vim
 │           ├─ gcc
 │           └─ python
```
systemd replaced most of this with declarative unit dependencies.


## Scheduling Overview

Scheduling is the operating system mechanism that decides which ready process gets the CPU next and for how long.

**Design Goals**:
- fairness
- high throughput
- low latency (interactive responsiveness)
- high CPU utilization
- predictable performance

**Basic Flow**  
`ready queue → scheduler selects → dispatch → run → block/preempt → back to queue`  
The scheduler uses a policy (FCFS, Round Robin, Priority, etc.) and a dispatcher to perform the context switch.

**Common policy examples**:
- FCFS: run in arrival order (simple, poor interactivity)
- Round Robin: time slices for fairness
- Priority: important jobs run first
- Multilevel / CFS (Linux): weighted fairness

## Scheduling Performance Metric

| Metric          | Measures            | Goal     |
| --------------- | ------------------- | -------- |
| Turnaround      | total job time      | minimize |
| Response        | first reaction time | minimize |
| Waiting         | queue delay         | minimize |
| Throughput      | work per second     | maximize |
| Mean completion | average turnaround  | minimize |
| Utilization     | CPU busy %          | maximize |
